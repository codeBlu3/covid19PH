```python
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import urlopen
```

```python
weburl = 'https://ourworldindata.org/coronavirus-testing-source-data'
```

```python
rawResponse = urlopen(weburl).read() 
```

```python
batSoup = BeautifulSoup(rawResponse)
```

```python
table = batSoup.find_all('table')[0]
```

```python
table_rows = table.find_all('tr')
```

```python
res = []
for tr in table_rows[1:]:
    td = tr.find_all('td')
    row = [tr.text.strip() for tr in td if tr.text.strip()]
    if row:
        res.append(row)
```

```python
worldTest = pd.DataFrame(res, columns=["Country", "Tests_total", "Date", "Source", "SourceDate", "Remarks"])
```
